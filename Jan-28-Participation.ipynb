```python
%%writefile code.js

function addTax(price, rate) {
  return price + (price*rate);
}

function createBurner(secret, limit) {
  let current = 1
  return function() {
    if (current <= limit) {
      current += 1
      return secret
    }
  }
}



const burner1 = createBurner("ABC", 3)

console.log(burner1())
console.log(burner1())
console.log(burner1())
console.log(burner1())
console.log(burner1())


console.log(addTax(10, 0.25))
console.log(addTax(10, 0.10))
```
